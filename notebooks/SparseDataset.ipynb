```python

num_rows = 50
num_features = 100
viable_features = 15
missing_features = num_features - viable_features


viable_data = np.random.rand(num_rows, viable_features) * 100


missing_data = np.empty((num_rows, missing_features))
missing_data[:] = np.nan


for col in range(missing_features):

    num_non_na = int(0.1 * num_rows)

    non_na_indices = np.random.choice(num_rows, num_non_na, replace=False)

    missing_data[non_na_indices, col] = np.random.rand(num_non_na) * 100


features = np.hstack((viable_data, missing_data))


_, target = make_classification(n_samples=num_rows, n_features=1, n_informative=1, n_redundant=0, n_clusters_per_class=1, weights=[0.5, 0.5], random_state=42)


columns = [f"feature_{i+1}" for i in range(num_features)]
df = pd.DataFrame(features, columns=columns)

df['target'] = target

df.to_csv('synthetic_classification_dataset_smallest.csv', index=False)
```
