```python
import numpy as np
import pandas as pd
```

```python
# Load the datasets
evapotranspiration = pd.read_csv("D:/DS & ML/CSVs/Kosakzi/Kosaksi-Pasapugazh-and-experiments/data/raw/Environment - ICRISAT/Evapotranspiration actual.csv")
evapotranspiration = evapotranspiration.drop(columns = ['Annual Evapotranspiration (mm)'])#optional
precipitation = pd.read_csv("D:/DS & ML/CSVs/Kosakzi/Kosaksi-Pasapugazh-and-experiments/data/raw/Environment - ICRISAT\Precipitation.csv")
precipitation = precipitation.drop(columns = ['Annual Precipitation (mm)'])#optional
fertilizer_consumption = pd.read_csv("D:/DS & ML/CSVs/Kosakzi/Kosaksi-Pasapugazh-and-experiments/data/raw/Environment - ICRISAT/Fertilizers consumption.csv")
```

```python
# Display the first few rows of each dataset
print("Evapotranspiration Data:")
evapotranspiration.head()
```

```python
print("\nPrecipitation Data:")
precipitation.head()
```

```python
print("\nFertilizer Consumption Data:")
fertilizer_consumption.head()
```

```python
cols1 = evapotranspiration.columns[5:]
cols2 = precipitation.columns[5:]
print(cols1,cols2,sep = "\n")
```

```python
# Calculate annual evapotranspiration by summing the monthly columns
evapotranspiration['annual_evapotranspiration'] = evapotranspiration[cols1].sum(axis=1)

# Calculate annual precipitation by summing the monthly columns
precipitation['annual_precipitation'] = precipitation[cols2].sum(axis=1)
```

```python
evapotranspiration.info()
```

```python
# Select relevant columns for merging
evapo_relevant = evapotranspiration[['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name', 'annual_evapotranspiration']]
precip_relevant = precipitation[['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name', 'annual_precipitation']]
fertilizer_relevant = fertilizer_consumption[['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name'] + [col for col in fertilizer_consumption.columns if col not in ['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name']]]
```

```python
# Merge datasets on common columns
merged_data = pd.merge(evapo_relevant, precip_relevant, on=['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name'], how='outer')
final_data = pd.merge(merged_data, fertilizer_relevant, on=['Dist Code', 'Year', 'State Code', 'State Name', 'Dist Name'], how='inner')
```

```python
# Display the first few rows of the final merged dataset
print("\nFinal Merged Data:")
final_data.head()
```

```python
# Save the final dataset to a new CSV file
final_data.to_csv('Combined_Data.csv', index=False)

print("\nCombined data has been saved to 'Combined_Data.csv'.")
```
