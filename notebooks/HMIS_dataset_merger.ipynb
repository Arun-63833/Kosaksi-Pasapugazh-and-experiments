```python
import os
import pandas as pd
import glob

base_dir = "/content/Kosaksi-Pasapugazh-and-experiments/data/raw/Health - HMIS"
all_data = []

for year in range(2008, 2017):
    year_folder = f"{year}-{year + 1}"
    sub_dir = os.path.join(base_dir, year_folder, year_folder)

    print(f"Processing year: {year_folder}")

    if os.path.exists(sub_dir):
        csv_files = glob.glob(os.path.join(sub_dir, "*.csv"))

        print(f"Found {len(csv_files)} CSV files in {sub_dir}")

        for file_path in csv_files:
            file_name = os.path.basename(file_path)
            state_name = file_name.replace(".csv", "")

            if state_name in ["All_India", "M O Defence", "M O Railways"]:
                continue

            try:
                df = pd.read_csv(file_path)

                df = df[df['Unnamed: 3'] == 'TOTAL']

                df = df.drop(columns=['Unnamed: 3'])

                df.columns = ['district', 'medical_code', 'medical_description', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december', 'january', 'february', 'march', 'total']

                df.insert(1, 'state', state_name)
                df.insert(0, 'year', year_folder)

                all_data.append(df)

            except Exception as e:
                print(f"Error processing file {file_name}: {e}")

if all_data:
    combined_data = pd.concat(all_data, ignore_index=True)
    combined_data.reset_index(drop=True, inplace=True)
    combined_data.to_csv("combined_entire_india_before_2017.csv", index=False)
    print("Data consolidation complete. Output saved as 'combined_entire_india_before_2017.csv'")
    print(f"Total rows in the combined data: {len(combined_data)}")
else:
    print("No data to concatenate. Please check your file paths and data.")
```

```python
import os
import pandas as pd
import glob

base_dir = "/content/Kosaksi-Pasapugazh-and-experiments/data/raw/Health - HMIS"
all_data = []

def get_total_column(month, columns):
    return next((col for col in columns if f'{month} Total' in col), None)

months = ['April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December', 'January', 'February', 'March', 'Total']

for year in range(2017, 2020):  # Changed year range to 2017 to 2020
    year_folder = f"{year}-{year + 1}"
    sub_dir = os.path.join(base_dir, year_folder, year_folder)

    print(f"Processing year: {year_folder}")

    if os.path.exists(sub_dir):
        csv_files = glob.glob(os.path.join(sub_dir, "*.csv"))

        print(f"Found {len(csv_files)} CSV files in {sub_dir}")

        for file_path in csv_files:
            file_name = os.path.basename(file_path)
            state_name = file_name.replace(".csv", "")

            if state_name in ["All_India", "M O Defence", "M O Railways"]:
                continue

            try:
                if file_path == "/content/Kosaksi-Pasapugazh-and-experiments/data/raw/Health - HMIS/2017-2018/2017-2018/Telangana.csv":

                    specific_headers = ['Data ItemWise Report for - Telangana', 'Unnamed: 1', 'Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4',
                                        'Unnamed: 9', 'Unnamed: 14', 'Unnamed: 19', 'Unnamed: 24',
                                        'Unnamed: 29', 'Unnamed: 34', 'Unnamed: 39', 'Unnamed: 44',
                                        'Unnamed: 49', 'Unnamed: 54', 'Unnamed: 59', 'Unnamed: 64']

                    try:
                        # Step 1: Read CSV file, skip first 5 rows and select specific columns
                        df = pd.read_csv(file_path,usecols=specific_headers, dtype=str)

                        # Step 2: Select rows where 'TOTAL' appears in Unnamed: 3 column
                        df = df[df['Unnamed: 3'].fillna('').str.contains('TOTAL', na=False)]

                        # Step 3: Drop Unnamed: 3 column
                        df = df.drop(columns=['Unnamed: 3'], errors='ignore')

                        # Step 4: Rename columns
                        df.columns = ['district', 'medical_code', 'medical_description', 'april', 'may', 'june', 'july', 'august', 'september', 'october', 'november', 'december', 'january', 'february', 'march', 'total']

                        # Step 5: Process district column
                        current_district = None
                        for index, row in df.iterrows():
                            if pd.notna(row['district']):
                                current_district = row['district']
                            else:
                                df.at[index, 'district'] = current_district

                        # Step 6: Insert state and year columns
                        df.insert(1, 'state', state_name)
                        df.insert(0, 'year', year_folder)

                        # Append processed dataframe to all_data list
                        all_data.append(df)

                    except Exception as e:
                        print(f"Error processing file {file_name}: {e}")

                else:
                    df = pd.read_csv(file_path, header=[0,1])

                    # Flatten the multi-index columns
                    df.columns = [' '.join(col).strip() for col in df.columns.values]

                    # Find the appropriate column names
                    dist_col = next((col for col in df.columns if 'Unnamed: 0' in col), None)
                    med_code_col = next((col for col in df.columns if 'Unnamed: 1' in col), None)
                    med_desc_col = next((col for col in df.columns if 'Unnamed: 2' in col), None)
                    stock_details_col = next((col for col in df.columns if 'Unnamed: 3' in col), None)

                    # Drop rows with stock details (keep only rows with 'TOTAL' in Unnamed: 3)
                    if stock_details_col:
                        df = df[df[stock_details_col].fillna('').str.contains('TOTAL', na=False)]

                    # Select only the relevant columns
                    columns_to_keep = [dist_col, med_code_col, med_desc_col] + [get_total_column(month, df.columns) for month in months]
                    df = df[columns_to_keep]

                    # Rename columns
                    df = df.rename(columns={
                        dist_col: 'district',
                        med_code_col: 'medical_code',
                        med_desc_col: 'medical_description'
                    })

                    for month in months:
                        total_col = get_total_column(month, df.columns)
                        if total_col:
                            df = df.rename(columns={total_col: month.lower()})


                    # Insert state and year columns
                    df.insert(0, 'year', year_folder)
                    df.insert(1, 'state', state_name)


                    # Rename Column names to lower case
                    df.columns = df.columns.str.lower()

                    # Remove rows where 'state' is in the state column
                    df = df[df['state'].str.lower() != 'state']

                    # Remove quotes from the start and end of medical_code
                    df['medical_code'] = df['medical_code'].str.strip("'")
                    all_data.append(df)

            except Exception as e:
                print(f"Error processing file {file_name}: {e}")

if all_data:
    combined_data = pd.concat(all_data, ignore_index=True)
    combined_data.reset_index(drop=True, inplace=True)
    combined_data.to_csv("combined_entire_india_2017_2020.csv", index=False)
    print("Data consolidation complete. Output saved as 'combined_entire_india_2017_2020.csv'")
    print(f"Total rows in the combined data: {len(combined_data)}")

else:
    print("No data to concatenate. Please check your file paths and data.")
```
